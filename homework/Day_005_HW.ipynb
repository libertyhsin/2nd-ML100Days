```python
#print(app_train.head())
print(app_train.describe())
app_train.corr()
```

```python
choose = ["AMT_CREDIT", "AMT_ANNUITY", "AMT_GOODS_PRICE"]
x = app_train["SK_ID_CURR"]
app_choose = app_train[choose]
print(app_choose.head())
print("---------SUM--------\n", app_choose.sum())
print("--------MEAN--------\n", app_choose.mean())
print("--------STD---------\n", app_choose.std())
```

```python
s = pd.Series(app_choose.loc[0:9, "AMT_CREDIT"], index=np.arange(0, 9))
s.plot()
```

```python
app_choose["AMT_CREDIT"].hist(bins=100)
app_choose.plot()
app_choose.cumsum().plot()
```

```python
ran = np.random.randn(10, 4)
print(ran)
print(ran.cumsum(0))
df = pd.DataFrame(ran.cumsum(0), columns=['A', 'B', 'C', 'D'], index=np.arange(0, 100, 10))
df.plot()
```
